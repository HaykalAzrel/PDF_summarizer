```python
# @title Import libraries
import os
import fitz
from google.colab import userdata
from langchain_community.llms import Replicate
from google.colab import files

#Set up Replicate API
api_token = userdata.get("api_token")
os.environ["REPLICATE_API_TOKEN"] = api_token

#Load IBM Granite model from Replicate
model_id = "ibm-granite/granite-3.3-8b-instruct"
llm = Replicate(model=model_id, replicate_api_token=api_token)
```

```python
# @title Upload a PDF file
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

# Extract text from the PDF
doc = fitz.open(pdf_path)
full_text = ""
for page in doc:
    full_text += page.get_text()

text_chunk = full_text[:3000]
```

```python
# @title Set custom parameters for model generation
parameters = {
    "top_k": 0,
    "top_p": 0.95,
    "max_tokens": 512,
    "min_tokens": 100,
    "random_seed": 42,
    "repetition_penalty": 1.1,
    "stopping_criteria": "length (512 tokens)",
    "stopping_sequence": None
}
```

```python
# @title Create summarization prompt
prompt = f"""
Berikut adalah isi dokumen. Tolong ringkas ke dalam 5 poin penting yang mencakup:
- Tujuan
- Isi utama
- Fakta penting
- Kesimpulan

{text_chunk}
"""
```

```python
# @title Summary using IBM Granite
summary = llm(prompt, model_kwargs=parameters)

# Print the summary
print("\n===== Ringkasan Dokumen =====\n")
print(summary)
```
